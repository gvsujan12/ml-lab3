```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_excel(r"C:\Users\gvsuj\Downloads\embeddingsdata.xlsx")
# Assuming you have two feature vectors, feature_vector1 and feature_vector2
# Replace these with your actual feature vectors
feature_v1 = df['embed_4'].to_numpy()
feature_v2 = df['embed_5'].to_numpy()

# Create an array of r values from 1 to 10
r_values = np.arange(1, 11)

# Initialize an empty list to store the calculated distances
distance_list = []

# Calculate Minkowski distance for each value of r
for r in r_values:
    distances = np.linalg.norm(feature_v1 - feature_v2, ord=r)
    distance_list.append(distances)
# Plot the distances vs. r
plt.plot(r_values, distance_list)
plt.xlabel('r Value')
plt.ylabel('Minkowski Distance')
plt.title('Minkowski Distance vs. r')
plt.grid(True)
plt.show()
```
