```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df=pd.read_excel(r"C:\Users\gvsuj\Downloads\embeddingsdata.xlsx")
# 'feature_1' is the feature you want to analyze
feature_1 = df['embed_5']  


# Creating a histogram
plt.hist(feature_1, bins=20)  # You can adjust the number of bins as needed
plt.xlabel('Feature Values')
plt.ylabel('Frequency')
plt.title('Histogram of Feature_of_Interest')
plt.show()

# Calculate the mean and variance
mean_feature_1= np.mean(feature_1)
variance_feature_1 = np.var(feature_1)

print("Mean of feature_1:" ,mean_feature_1)
print("Variance of feature_1: ",variance_feature_1)
```
